```python
# 📚 Step 1: Importing the pandas library
# A "library" in Python is like a toolkit.
# Imagine you’re a carpenter. Instead of making your own hammer and screwdriver every time,
# you open a toolbox (library) that already has all the tools ready to use.
# Pandas is a toolbox for working with tables of data (like Excel).
import pandas as pd  # We call it 'pd' to save typing later
```

```python
# 📄 Step 2: Reading the data from a CSV file
# "read_csv" is like opening an Excel file that has rows and columns.
# This file has the Age and Height of students.
# We load it into Python so we can look at it, work with it, and even make predictions later.
data = pd.read_csv("Age_Height_Dataset.csv")
```

```python
# 👀 Step 3: Looking at the first 5 rows of the data
# This helps us quickly see what’s inside the dataset.
# It's like flipping to the first page of a notebook to check what kind of notes are written.
data.head()
```

```python
# 🖍️ Step 1: Import the drawing tool
# matplotlib is a library that helps us draw charts and graphs.
# pyplot is the part of it that lets us draw things like points and lines.
import matplotlib.pyplot as plt

# 🟦 Step 2: Draw dots (scatter plot) for each student's age and height
# Each dot shows one student's age and height.
# X-axis: Age → how old the student is
# Y-axis: Height → how tall the student is
plt.scatter(data['Age'], data['Height'], color='blue')

# 🏷️ Step 3: Give a title to the graph
plt.title('Age vs. Height of Students')

# 🧭 Step 4: Label the X-axis (bottom of the graph)
plt.xlabel('Age (years)')

# 🧭 Step 5: Label the Y-axis (side of the graph)
plt.ylabel('Height (cm)')

# ➕ Step 6: Add a grid to make reading the chart easier
plt.grid(True)

# 👀 Step 7: Show the graph
# This will pop up the chart with all the dots
plt.show()
```

```python
# 📦 Step 1: Import the tools we need
# We are using sklearn – a library that helps us do machine learning easily.
from sklearn.linear_model import LinearRegression
import numpy as np

# 🎯 Step 2: Define our input (X) and output (y)
# X is the known value (Age), and y is what we want to predict (Height)
X = data[['Age']]        # Double brackets [[]] because sklearn needs 2D input
y = data['Height']    # The height column is our output

# 🤖 Step 3: Create the Linear Regression model
# This is like building a robot that will learn the pattern between age and height
model = LinearRegression()

# 📚 Step 4: Teach the model using our data
# We give it the age and height examples, and it learns the trend
model.fit(X, y)

# 🔮 Step 5: Use the model to make predictions
# Now we ask: “If someone is 13 years old, what height should we expect?”
predicted_height = model.predict([[13]])

# 🖨️ Step 6: Show the result
print(f"Predicted height for a 13-year-old: {predicted_height[0]:.2f} cm")
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd

# Step 1: Load the dataset (already created earlier)
data = pd.read_csv("Age_Height_Dataset.csv")

# Step 2: Split into input (X) and output (y)
X = data[['Age']]           # X should be 2D
y = data['Height']       # y is the height

# Step 3: Create and train the model
reg = LinearRegression()
reg.fit(X, y)

# Step 4: Get slope (m) and intercept (b)
m = reg.coef_[0]            # slope
b = reg.intercept_          # intercept

print(f"Slope (m): {m:.2f}")
print(f"Intercept (b): {b:.2f}")
```

```python
# Predict height manually using the formula: y = m * x + b
age_input = 13
predicted_height = m * age_input + b

print(f"Predicted height for age {age_input} is {predicted_height:.2f} cm")
```
